```
import os
import cv2
import dlib
import numpy
import imutils
from imutils.face_utils import rect_to_bb
from imutils.face_utils import FaceAligner
from matplotlib import pyplot
```

```
BASE_PATH = "drive/My Drive/FER/datasets/fer_aligned/"
DATA_PATH = "fer/final/"
```

```
face_detector = dlib.get_frontal_face_detector()
shape_predictor = dlib.shape_predictor("drive/My Drive/FER/Colab/GauravSharma/Multi-Input-Model/dumps/shape_predictor_68_face_landmarks.dat")

desiredFaceWidth = 48
fa = FaceAligner(shape_predictor, desiredLeftEye=(0.32, 0.32), desiredFaceWidth=desiredFaceWidth)
```

```
for dir_ in os.listdir(DATA_PATH):
    print(f"processing {dir_} images...")

    for f in os.listdir(DATA_PATH + dir_ + "/"):
        img = cv2.imread(DATA_PATH + dir_ + "/" + f, 0)
        faces = face_detector(img, 2)

        if faces:
            for face in faces:
                try:
                    (x, y, w, h) = rect_to_bb(face)
                    # faceOrig = imutils.resize(img[y:y + h, x:x + w], width=desiredFaceWidth)
                    faceAligned = fa.align(img, img, face)
                    cv2.imwrite(BASE_PATH+dir_.title()+"/"+f, faceAligned)

                    # pyplot.imshow(img, cmap="gray")
                    # pyplot.show()
                    # pyplot.imshow(faceAligned, cmap="gray")
                    # pyplot.show()
                except Exception:
                    pass
        else:
            continue
```

```
total_aligned_images = 0

for dir_ in os.listdir(BASE_PATH):
    count = 0
    for f in os.listdir(BASE_PATH + dir_ + "/"):
        count += 1
    print(f"{dir_} has {count} aligned images")
    total_aligned_images += count

print(f"total aligned images {total_aligned_images}")
```
